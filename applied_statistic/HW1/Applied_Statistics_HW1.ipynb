```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
bacterias = pd.read_csv('Protein_lengths_bacterias.csv')
eucaryotes = pd.read_csv('Protein_lengths_eucaryotes.csv')
```

```python
bacterias
```

```python
eucaryotes
```

```python
data = pd.concat([bacterias['Length.2'], eucaryotes['Length']], axis=1).to_numpy()
data
```

```python
max(bacterias['Length.2'])
```

```python
max(eucaryotes['Length'])
```

```python
plt.clf()
colors = ['blue', 'lime']
plt.hist(data, alpha = 0.7, color = colors, label = ['bacterium', 'eukaryotes'], histtype ='bar')
plt.title('Совместная гистограмма, автоматический шаг',
          fontweight = "bold")
plt.xlabel('длина белка')
plt.ylabel('количество')
plt.legend(prop ={'size': 10})
plt.savefig('mutual_auto.pdf')
plt.show()
```

```python
plt.clf()
colors = ['blue', 'lime']
plt.hist(data, bins=int(data.shape[0] ** 0.5), alpha = 0.7, color = colors, label = ['bacterium', 'eukaryotes'], histtype ='bar')
plt.title('Совместная гистограмма, шаг $[\sqrt{n}]$',
          fontweight = "bold")
plt.xlabel('длина белка')
plt.ylabel('количество')
plt.legend(prop ={'size': 10})
plt.savefig('mutual_sqrt_n.pdf')
plt.show()
```

```python
plt.clf()
colors = ['blue', 'lime']
plt.hist(data, bins=int(np.log2(data.shape[0]) + 1), alpha = 0.7, color = colors, label = ['bacterium', 'eukaryotes'], histtype ='bar')
plt.title('Совместная гистограмма, шаг $[log_2n+1]$',
          fontweight = "bold")
plt.xlabel('длина белка')
plt.ylabel('количество')
plt.legend(prop ={'size': 10})
plt.savefig('mutual_sturges.pdf')
plt.show()
```

```python
plt.clf()
colors = ['blue', 'lime']
plt.hist(data, bins=int(2 * data.shape[0] ** (1/3)), alpha = 0.7, color = colors, label = ['bacterium', 'eukaryotes'], histtype ='bar')
plt.title('Совместная гистограмма, шаг $[2\sqrt[3]{n}]$',
          fontweight = "bold")
plt.xlabel('длина белка')
plt.ylabel('количество')
plt.legend(prop ={'size': 10})
plt.savefig('mutual_rice.pdf')
plt.show()
```

```python
plt.clf()
bplot1 = plt.boxplot([bacterias['Length.2'], eucaryotes['Length']],
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=['bacterium', 'eukaryotes'])  # will be used to label x-ticks

for patch, color in zip(bplot1['boxes'], colors):
        patch.set_facecolor(color)
plt.title('Box plot длин белков', fontweight = "bold")
plt.xlabel('тип организма')
plt.ylabel('длина белка')
plt.savefig('boxplot.pdf')
plt.show()
```

```python
plt.clf()
sns.ecdfplot(data=data, palette=['blue', 'lime'])
plt.legend(['eukaryotes', 'bacterium'])
plt.ylabel('пропорция')
plt.xlabel('длина белка')
plt.title('CDF', fontweight = "bold")
plt.savefig('cdf.pdf')
plt.show()
```
