```python
x = [i*600 for i in range(1, 9)]
```

```python
t_c = np.loadtxt('times_cpp.txt')
gf_c = np.loadtxt('flops_cpp.txt')
t_py = np.loadtxt('times_py.txt') * 1000
gf_py = np.loadtxt('flops_py.txt')
t_cy = np.loadtxt('alex_times_cython.txt') * 1000
gf_cy = np.loadtxt('alex_gflops_cython.txt') / 1e9
```

```python
plt.plot(x, t_c[:-2], label='C++')
# plt.plot(x, t_py[:-2])
plt.plot(x, t_cy[:-2], label='Cython')
plt.xlabel('Array size')
plt.ylabel('Time in ms')
plt.legend()
plt.savefig("c++_cython.png", bbox_inches="tight")
plt.show()

plt.plot(x, t_py[:8], label='Python')
plt.xlabel('Array size')
plt.ylabel('Time in ms')
plt.legend()
plt.savefig("python.png", bbox_inches="tight")
plt.show()
```

```python
plt.plot(x, gf_c[:-2], label='C++')
plt.plot(x, gf_cy[:8], label='Cython')
plt.xlabel('Array size')
plt.ylabel('Giga flops')
plt.legend()
plt.savefig("gf_c++_cython.png", bbox_inches="tight")
plt.show()

plt.plot(x, gf_py[:8], label='Python')
plt.xlabel('Array size')
plt.ylabel('Mega flops')
plt.legend()
plt.savefig("gf_python.png", bbox_inches="tight")
plt.show()
```
