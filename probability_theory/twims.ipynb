```python
import scipy.stats as st
from scipy.stats import binom
from scipy.stats import cauchy
import matplotlib.pyplot as plt
import math

datalist = []
for i in range(10000):
  xi = binom.rvs(n=1, p=0.5, loc=0, size=i+1)
  #xi = cauchy.rvs(size=i+1)
  temp = sum(xi) / (i+1)
  datalist.append(temp)
plt.clf()
plt.plot(datalist, '-g', label='Graphic', alpha=0.5)
plt.legend(loc='best')
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import math

data = yf.download('AAPL', '2016-01-01', '2019-01-01')
data['Adj Close'].plot()

fig, ax = plt.subplots()
logReturn = []
for i in range(len(data['Adj Close'])-1):
  logReturn.append(math.log(data['Adj Close'][i+1] / data['Adj Close'][i]))
#ax.hist(logReturn)

import pandas as pd
print(pd.Series(logReturn).describe())
plt.boxplot(logReturn)
```
